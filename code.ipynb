```python
# Binear Search
def binarySearch(arr, target):
    l = 0
    r = len(arr) - 1

    while l <= r:
        mid = (r-l)//2

        if target > arr[mid]:
            l = mid + 1
        elif target < [mid]:
            r = mid - 1
        else:
            return mid
    return -1
```

```python
def isCorrect(val):
    if val > 10:
        return 1
    if val < 10:
        return -1
    return 0 

def binarySearchRange(low, high, target):
    while low <= high:
        mid = (high-low)//2
        if isCorrect(mid) < 0:
            high = mid - 1
        elif isCorrect(mid) > 0:
            low = mid + 1
        else:
            return mid
    return -1
```

```python
class Node:
    def __init__(self, val, left=None, right=None) -> None:
        self.val = val
        self.left = left
        self.right = right

def bst_search(root: Node, target):
    if not root:
        return False
    
    if target > root.val:
        return bst_search(root.right, target)
    elif target < root.val:
        return bst_search(root.left, target)
    else:
        return True
```

```python
class Node:
    def __init__(self, val, left=None, right=None) -> None:
        self.val = val
        self.left = left
        self.right = right

class BST:
    def __init__(self):
        self.root = None

    def insert(self, val, node: Node=None):
        if not self.root:
            self.root = Node(val)
            return 
        
        if not node:
            node = self.root
        if val > node.val:
            node.right = self.insert(val, node.right)
        elif val < node.val:
            node.left = self.insert(val, node.left)

    def remove(self, val, node: Node=None):
        if not self.root:
            return None
        
        if not node:
            node = self.root

        if val > node.val:
            node.right = self.remove(node.right, val)
        elif val < node.val:
            node.left = self.remove(node.left, val)
        else:
            min_node = self._min_node(node.right)
            node.val = min_node.val
            node.right = self.remove(node.right, min_node.val)

    def _min_node(self, node: Node):
        curr = node
        while curr and curr.left:
            curr = curr.left
        return curr
```

```python
def inorder_dfs(root: Node):
    if not root:
        return
    
    inorder_dfs(root.left)
    # do something with the value
    inorder_dfs(root.right)

def preporder(root: Node):
    if not root:
        return
    
    #do something with the value
    preporder(root.left)
    preporder(root.right)

def postorder(root: Node):
    if not root:
        return
    
    postorder(root.left)
    postorder(root.right)
    #do something with the value
```

```python
from collections import deque

def bfs(root: Node):
    queue = deque()

    if root:
        queue.append(root)

    level = 0
    while queue:
        # do something at level
        for i in range(len(queue)):
            curr = queue.popleft()
            # do something at node
            if curr.left:
                queue.append(curr.left)
            if curr.right:
                queue.append(curr.right)
        level += 1
```

```python
path = []
def backtrack_path(root: Node):
    if not root or root.val == 0: # 0 being not viable path
        return False
    path.append(root.val)

    if not root.left and not root.right: # a valid path
        return True
    if backtrack_path(root.left):
        return True
    if backtrack_path(root.right):
        return True
    path.pop()
    return False
```

```python
class MinHeap:
    def __init__(self):
        self.heap = [-1]

    def get_min(self) -> int:
        if len(self.heap) > 1:
            return self.heap[1]
        return None

    def push(self, val):
        self.heap.append(val)
        i = len(self.heap) - 1

        while self.heap[i] < self.heap[i//2]:
            tmp = self.heap[i]
            self.heap[i] = self.heap[i//2]
            self.heap[i//2] = tmp
            i = i // 2

    def pop(self):
        if len(self.heap) == 1: # because we're starting at 1 instead of 0
            return None
        if len(self.heap) == 2:
            self.heap.pop()

        head = self.heap[1]
        i = 1
        while 2 * i < len(self.heap): #still have children
            left_i = 2 * i
            right_i = left_i + 1

            if all([
                right_i < len(self.heap),
                self.heap[right_i] < self.heap[left_i],
                self.heap[i] > self.heap[right_i]
            ]):
                tmp = self.heap[i]
                self.heap[i] = self.heap[right_i]
                self.heap[right_i] = tmp
                i = right_i
            elif self.heap[i] > self.heap[left_i]:
                tmp = self.heap[i]
                self.heap[i] = self.heap[left_i]
                self.heap[left_i] = tmp
                i = left_i
            else:
                break

        return head


    def heapify(self, arr: list):
        arr.append(arr[0])
        self.heap = arr
        curr = (len(self.heap)-1)//2

        while curr > 0:
            i = curr
            left_i = 2 * i
            right_i = left_i + 1

            while left_i < len(self.heap): #still have children
                if all([
                    right_i < len(self.heap),
                    self.heap[right_i] < self.heap[left_i],
                    self.heap[i] > self.heap[right_i]
                ]):
                    tmp = self.heap[i]
                    self.heap[i] = self.heap[right_i]
                    self.heap[right_i] = tmp
                    i = right_i
                elif self.heap[i] > self.heap[left_i]:
                    tmp = self.heap[i]
                    self.heap[i] = self.heap[left_i]
                    self.heap[left_i] = tmp
                    i = left_i
                else:
                    break
            curr -= 1
```

```python
from typing import Any
class Pair:
    def __init__(self, key, val):
        self.key = key
        self.val = val

class HashMap:
    '''Open Addressing'''
    def __init__(self, capacity=2) -> None:
        self.size = 0
        self.capacity = capacity
        self.map: list[Pair] = [None for _ in range(capacity)]

    def hash(self, key: str):
        # any hashing function here
        index = 0
        for c in key:
            index += ord(c)
        return index & self.capacity
    
    def get(self, key: str) -> Any:
        index = self.hash(key)

        while self.map[index] != None:
            if self.map[index].key == key:
                return self.map[index].val
            index += 1
            index = index % self.capacity
        return None
    
    def put(self, key, val):
        index = self.hash(key)

        while True:
            if self.map[index] == None: # empty and good to place
                self.map[index] = Pair(key, val)
                self.size += 1
                if self.size >= self.capacity // 2: # is the capacity too small and going to cause too much collision
                    self.rehash()
                return
            elif self.map[index].key == key: # key already exists
                self.map[index].val = val
                return
            
            index += 1
            index = index % self.capacity

    def remove(self, key):
        if not self.get(key):
            return
        
        index = self.hash(key)
        while True:
            if self.map[index].key == key:
                self.map[index] = None
                # this will cause a bug with get because it leaves a hole
                self.size -= 1
                return
            index += 1
            index = index % self.capacity

    def rehash(self):
        self.capacity *= 2
        new_map = [None for _ in range(self.capacity)]
        old_map = self.map
        self.size = 0

        for pair in old_map:
            if pair:
                self.put(pair.key, pair.val)
```

```python
def matrix_dfs_n_paths(grid, r, c, visited=-1):
    ROWS = len(grid)
    COLS = len(grid[0]) #assuming uniform grid

    if any([
        r == ROWS,
        c == COLS,
        (r, c) in visited,
        grid[r][c] == 1,
    ]):
        return 0
    
    if r == ROWS - 1 and c == COLS - 1:
        return 1
    
    if visited == -1:
        visited = set()
    visited.add((r,c))

    count = 0 
    neighbors = [(1,0), (0,1), (-1,0), (0,-1)]
    for dr, dc in neighbors:
        count += matrix_dfs_n_paths(grid, r+dr, c+dc, visited)
    visited.remove((r, c))

    return count
```

```python
from collections import deque

def matrix_bfs_shortest_path(grid):
    ROWS = len(grid)
    COLS = len(grid[0]) #assuming uniform grid
    visited = set()
    queue = deque()
    visited.add((0,0))

    length = 0
    while queue:
        for _ in range(len(queue)):
            r, c = queue.popleft()
            if r == ROWS - 1 and c == COLS - 1:
                return length
            
            neighbors = [(1,0), (0,1), (-1,0), (0,-1)]
            for dr, dc in neighbors:
                if any([
                    min((r+dr, c+dc)) < 0,
                    r + dr == ROWS,
                    c + dc == COLS,
                    (r+dr, c+dc) in visited,
                    grid[r+dr][c+dc] == 1,
                ]):
                    continue
                
                queue.append(r+dr, c+dc)
                visited.add((r+dr, c+dc))
        length += 1
```

```python
def fibinacci_brute_force(n):
    if n <= 1: 
        return n
    return fibinacci_brute_force(n-1) + fibinacci_brute_force(n-2)

cache = {}
def fibinacci_top_down(n):
    if n <= 1:
        return n
    if n in cache:
        return cache[n]
    
    cache[n] = fibinacci_top_down(n-1) + fibinacci_top_down(n-2)
    return cache[n]

def fibinacci_bottom_up(n):
    if n <= 1:
        return n
    
    dp = [0, 1]
    i = 2
    while i <= n:
        tmp = dp[1]
        dp[1] = dp[0] + dp[1]
        dp[0] = tmp
        i += 1
    return dp[1]
```
